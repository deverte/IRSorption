## Поиск пиков

### Десорбция

```python
processing.interval_analysis(equalized, 100, 600, mode=3, extrema="min", output="Full", directory=directory)
```

При десорбции, видно два пика температуры. На образцах $SiO_2$ и $Au/SiO_2$ эти пики более заметны чем на образцах $Al_2O_3$ и $Au/Al_2O_3$ и находятся практически на одном уровне (разница температур примерно 0.1 $^oC$) и время, которое прошло между их проявлением - около 100 с. При чем первый пик имеет меньшую температуру.

На образцах $Al_2O_3$ и $Au/Al_2O_3$ первый пик практически не заметен, а второй значительно выделяется и имеет температуру, меньшую чем минимум на $SiO_2$ и $Au/SiO_2$ (разница примерно на 1.1 $^oC$).

1. Возможные объяснения двух экстремумов.

(1.1.) Можно сделать следующее предположение - на образцах изначально имеются молекулы $H_2O$ и они десорбируются порциями. Данный эффект может быть обусловлен тем, что молекулы воды могут быть с другими примесями более тяжелых, или наоборот легких элементов, что "закупоривает" поры материалов.

(1.2.) Так же можно сделать другое предположение - т.к. образцы имеют пористую структуру, возможно существуют "зоны", в которых нужно приложить меньшую силу для преодоления межмолекулярного взаимодействия.

2. Возможные объяснения разницы в проявлении двух экстремумов.

(2.1.) Согласно первому возможному объяснению двух экстремумов (1.1.), на образцах $SiO_2$ и $Au/SiO_2$ сначала могут десорбироваться молекулы воды с примесями, затем либо снова адсорбироваться те же молекулы, либо адсорбироваться другие, более легкие молекулы или атомы (что может объяснить кратковременный рост после первого минимума) (адсорбироваться так же могут и молекулы воды с соседних образцов), а затем снова десорбироваться молекулы, но уже окончательно. А на образцах $Al_2O_3$ и $Au/Al_2O_3$ из-за другой структуры, молекулы не могут снова десорбироваться, что дает возможность более активно десорбироваться другим, "более глубоким" молекулам.

(2.2.) Согласно второму возможному объяснению двух экстремумов (1.2.), векторное поле сил межмолекулярного взаимодействия, может иметь различный вид для разных образцов и напрямую зависеть от их геометрии.

### Адсорбция (атмосферный воздух)

```python
processing.interval_analysis(equalized, 950, 1080, mode=3, extrema="max", output="Full", directory=directory)
```

При адсорбции атмосферного воздуха, на образце $Au/SiO_2$ заметно два пика, что может быть связано с подобным эффектом при десорбции. Но на других графиках подобного поведение не видно, по крайней мере не так явно.
