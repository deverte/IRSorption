```python
import pandas as pd
import numpy as np
import normalization
```

# Нормировка

## Получение данных

```python
fileIn = "Data/e_1_SiO2_Al2O3.dat" # Входной файл
fileOut = "Data/n_1_SiO2_Al2O3.dat" # Выходной файл
```

Считываем данные из файла fileIn.

```python
data = pd.read_csv(fileIn, index_col=0, dtype=np.double)
data.head()
```

```python
data.plot();
```

## Нормировка

```python
normalized = normalization.normalize(data, 100, 600)
normalized.plot();
```

## Экспорт данных

Экспортируем призводные от данных в файл fileOut.

```python
normalized.to_csv(fileOut)
```
